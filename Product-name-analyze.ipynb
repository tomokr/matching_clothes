## Age group detection

```python
def age_group_extract(df):
    y = re.search('^Baby', df['0_name'])
    if y == None:
        df['age_group'] = 'Toddler'
    else:
        df['age_group'] = 'Baby'
    return df
```

```python
df_9_12m_group = df_9_12m.apply(age_group_extract, axis = 1)
```

```python
df_5t_group = df_5t.apply(age_group_extract, axis=1)
```

# Using word vectorization for finding the common word

```python
list_9_12m = df_9_12m_group['0_name'].tolist()
```

```python
vectorizer = CountVectorizer(list_9_12m,max_df=0.30) #remove boys/girls/baby/carter's
X = vectorizer.fit_transform(list_9_12m)
```

```python
print(vectorizer.get_feature_names())
```

```python
no_mean_list = ['12m', '2pc', '2pk', '3pc', '3pk', '4pc', '4pk', '6pk', '9m', 'and', 'with', 'the', 'toddler', 'amp', 'little','on']
color_list = ['black', 'blue','gray', 'green', 'pink', 'purple', 'red', 'white', 'yellow', 'orange','brown','navy','aqua','coral',
              'gold', 'heather','mint','olive','sage']
fashion_list = ['bodysuit', 'bottom', 'cardigan', 'dress', 'dresses', 'footed','hat', 'jumpsuit',  'piece', 'print', 
                'printed',  'romper', 'shirt', 'short', 'sleeve', 'snowsuit', 'socks', 'swimsuit',  'top', 'pajama', 'long', 'rash',
                'sleep', 'pants','striped','play', 'ankle','basic','chambray','chenille','converter','pull','dressy','nightgown','pull','ruffle','skirtall','sunsuit']

remove_list = no_mean_list + color_list+fashion_list
```

```python
len(vectorizer.get_feature_names())
```

```python
count_vect_df = pd.DataFrame(X.todense(), columns=vectorizer.get_feature_names())
```

```python
count_vect_df = count_vect_df.drop(remove_list, axis=1)
```

```python
df_9_12m_group = df_9_12m_group.drop(['Unnamed: 0'], axis=1)
```

```python
df_9_12m_word = df_9_12m_group.join(count_vect_df)
```

```python
def generate_text_from_df(df):
    text = ""
    for index, row in df.iterrows():
        target_url = 'https://www.target.com'
        text += row['0_name'] + "<>" + target_url + row['0_link'] + "<>"
    return text
```

```python
generate_text_from_df(df_output)
```

# Generating HTML file

```python
def generate_html_from_df(df):
    target_url = 'https://www.target.com'
    df['html_string'] = '<a href="'+target_url+df['0_link']+'" target="_blank">'+df['0_name']+'</a>'
    return df
```

```python
df_output_html = df_output.apply(generate_html_from_df, axis = 1)
```

```python
df_output_list = df_output_html['html_string'].values.tolist()
```

```python
def listToString(s): 
    str1 = "" 
    for ele in s: 
        str1 += ele+"<br>"
    return str1
```

```python
listToString(df_output_list)
```

```python
html = "<h1>5t/9-12m</h1> <h2>keywords</h2>"
for index, row in sum_df_mod.iterrows():
    html += '<a href="#'+row["index"] + '">' + row["index"] +'</a> '
for index, row in sum_df_mod.iterrows():
    html += '<h2 id="'+row["index"] + '">' + row["index"] +'</h2>'
    df_output = df_5t_and_9_12m[df_5t_and_9_12m[row['index']]==1]
    df_output_html = df_output.apply(generate_html_from_df, axis = 1)
    df_output_list = df_output_html['html_string'].values.tolist()
    html += listToString(df_output_list)
```

```python
with open('output.html', mode='w') as f:
    f.write(html)
```
