```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import pandas as pd
import re
```

# Load data

```python
df_9_12m = pd.read_csv('df_9_12m.csv')
```

```python
df_5t = pd.read_csv('df_5t.csv')
```

# Product name pattern
Toddler/Baby Girls'/Boys'/"none" "design" Dress/T-shirt/Romper/Bodysuit.. - Just One You® made by carter\'s 

## Age group detection

```python
def age_group_extract(df):
    y = re.search('^Baby', df['0_name'])
    if y == None:
        df['age_group'] = 'Toddler'
    else:
        df['age_group'] = 'Baby'
    return df
```

```python
df_9_12m_group = df_9_12m.apply(age_group_extract, axis = 1)
```

```python
df_9_12m_group
```

# Using word vectorization for finding the common word

```python
list_9_12m = df_9_12m_group['0_name'].tolist()
```

```python
vectorizer = CountVectorizer(list_9_12m,max_df=0.30) #remove boys/girls/baby/carter's
X = vectorizer.fit_transform(list_9_12m)
```

```python
print(vectorizer.get_feature_names())
```

```python
len(vectorizer.get_feature_names())
```

```python
count_vect_df = pd.DataFrame(X.todense(), columns=vectorizer.get_feature_names())
```

```python
count_vect_df.drop(['12m', '2pc', '2pk', '3pc', '3pk', '4pc', '4pk', '6pk', '9m'], axis=1)
```

```python
df_9_12m_group = df_9_12m_group.drop(['Unnamed: 0'], axis=1)
```

```python
df_9_12m_word = df_9_12m_group.join(count_vect_df)
```

```python
df_9_12m_word
```

```python
list_5t = df_5t['0_name'].tolist()
```

```python
Y = vectorizer.transform(list_5t)
```

```python
count_5t_df = pd.DataFrame(Y.todense(), columns=vectorizer.get_feature_names())
```

```python
count_5t_df.drop(['12m', '2pc', '2pk', '3pc', '3pk', '4pc', '4pk', '6pk', '9m'], axis=1)
```

```python
df_5t = df_5t.drop(['Unnamed: 0'], axis=1)
```

```python
df_5t_word = df_5t.join(count_5t_df)
```

# Search the common word in 5t and 9-12m

```python
df_5t_and_9_12m = pd.concat([df_5t_word, df_9_12m_word])
```

```python
df_5t_and_9_12m = df_5t_and_9_12m.drop(['12m', '2pc', '2pk', '3pc', '3pk', '4pc', '4pk', '6pk', '9m'], axis=1)
```
