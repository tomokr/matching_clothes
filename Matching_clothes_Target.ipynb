```python
import pandas as pd
import re
```

```python
f = open('result_babies_6-9m.txt', 'r')
```

```python
lines = [line for line in f.readlines()]
f.close()
lines
```

```python
lines[0]
```

```python
def product_title_extract(string):
    x = re.search('sametab">(.*)</a>', string)
    return x.group(1)
```

```python
product_title_array = map(product_title_extract, lines)
product_title_list = list(product_title_array)
product_title_list
```

```python
product_babies_all = pd.DataFrame.from_dict(product_title_list)
```

```python
f = open('result_5t.txt', 'r')
lines_5t = [line for line in f.readlines()]
f.close()
```

```python
product_title_array = map(product_title_extract, lines_5t)
product_title_list = list(product_title_array)
product_5t_all = pd.DataFrame.from_dict(product_title_list)
```

```python
product_5t_all
```

```python
target_url = 'https://www.target.com'
```

```python
def product_link_extract(string):
    y = re.search('href="(.*)#lnk', string)
    return y.group(1)
```

```python
product_link_array = map(product_link_extract, lines)
product_link_list = list(product_link_array)
product_baby_link = pd.DataFrame.from_dict(product_link_list)
```

```python
product_babies_all_ = product_babies_all.join(product_baby_link, lsuffix='_name', rsuffix='_link')
```

```python
product_babies_all_
```

```python
product_link_array = map(product_link_extract, lines_5t)
product_link_list = list(product_link_array)
product_5t_link = pd.DataFrame.from_dict(product_link_list)
```

```python
product_5t_all_ = product_5t_all.join(product_5t_link, lsuffix='_name', rsuffix='_link')
```

```python
product_5t_all_
```

```python
product_babies_and_5t = pd.merge(product_babies_all_, product_5t_all_, on='0_name', how='inner')
```

```python
product_babies_and_5t
```
